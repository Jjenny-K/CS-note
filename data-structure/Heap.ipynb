### Heap structure - Example Code

#### max-heap
- parent(i) = child(i) // 2
- left_child(i) = parent(i) * 2
- right_child(i) = (parent(i) * 2) + 1

```python
class max_heap:
    def __init__(self, value):
        # root node index '1'로 초기화
        self.heap = [None, value]

    def parent(self, pos):
        return pos // 2

    def left_child(self, pos):
        return pos * 2

    def right_child(self, pos):
        return (pos * 2) + 1

    def swap(self, fpos, spos):
        self.heap[fpos], self.heap[spos] = self.heap[fpos], self.heap[spos]

    def print(self):
        print(self.heap)

    def insert(self, value):
        # 마지막 노드 추가
        # max-heap을 유지할 수 있도록 부모 노드와 위치 재정렬
        if len(self.heap) < 1:
            self.heap = [None, value]
            return

        self.heap.append(value)

        insert_index = len(self.heap) - 1

        while self.heap[insert_index] > self.heap[self.parent(insert_index)]:
            self.swap(insert_index, self.parent(insert_index))
            insert_index = self.parent(insert_index)

    def chk_delete_swap(self, pos):
        # left_child, right_child와 비교해 재정렬 유무 반환
        if len(self.heap) > self.left_child(pos):
            if len(self.heap) > self.right_child(pos):
                return self.heap[pos] < max(self.heap[self.left_child(pos)], self.heap[self.right_child(pos)])
            else:
                return self.heap[pos] < self.heap[self.left_child(pos)]
        else:
            return False

    def delete_use_cds(self):
        # 첫번째 노드 삭제
        # 마지막 노드를 첫번째 노드로 이동한 후 max-heap을 유지할 수 있도록 위치 재정렬
        if len(self.heap) <= 1:
            return

        print('delete node:', self.heap[1])
        
        self.heap[1] = self.heap[-1]
        self.heap.pop()

        change_index = 1

        while self.chk_delete_swap(change_index):
            if len(self.heap) > self.right_child(change_index):
                if self.heap[self.left_child(change_index)] > self.heap[self.right_child(change_index)]:
                    self.swap(change_index, self.left_child(change_index))
                    change_index = self.left_child(change_index)
                else:
                    self.swap(change_index, self.right_child(change_index))
                    change_index = self.right_child(change_index)
            else:
                self.swap(change_index, self.left_child(change_index))
                change_index = self.left_child(change_index)

    def max_heapify(self, pos):
        # 항상 max-heap을 유지할 수 있도록 주어진 노드 위치 재정렬
        left_index = self.left_child(pos)
        right_index = self.right_child(pos)
        max_index = pos

        if left_index <= (len(self.heap) - 1) and self.heap[max_index] < self.heap[left_index]:
            max_index = left_index
        
        if right_index <= (len(self.heap) - 1) and self.heap[max_index] < self.heap[right_index]:
            max_index = right_index

        if max_index != pos:
            self.swap(pos, max_index)
            self.max_heapify(max_index)

    def delete_use_mh(self):
        # 첫번째 노드 삭제
        # 마지막 노드를 첫번째 노드로 이동한 후 max-heap을 유지할 수 있도록 위치 재정렬
        if len(self.heap) <= 1:
            return

        print('delete node:', self.heap[1])
        
        self.heap[1] = self.heap[-1]
        self.heap.pop()
        
        self.max_heapify(1)
```
